```python
peter_check = []
candidates = []
for i in  range(len(mota_list)):
    relevant_top_5 = my_search.search_top_down(mota_list[i])
    print(relevant_top_5)
    tmp = []
    for relevant in relevant_top_5:
        tmp.append(int(my_search.document_4_so[relevant].split(' *** ')[0]))
    print(tmp)
    candidates.append(tmp)
    hscode = int(str(hscode_list[i])[:4])
    print(hscode)
    if hscode in tmp:
        peter_check.append('True')
    else : peter_check.append('False')
    # break
```
